```python
import torch
import torchvision

image_path = './'

celeba_dataset = torchvision.datasets.CelebA(
    image_path, split='train', target_type='attr', download=True
)
```

```python

assert isinstance(celeba_dataset, torch.utils.data.Dataset)
```

```python
example = next(iter(celeba_dataset))

print(example)
```

```python
from itertools import islice
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15, 10))

for i, (img, labels) in islice(enumerate(celeba_dataset), 18):
    ax = fig.add_subplot(3, 6, i+1)
    ax.imshow(img)
    ax.set_title(labels[31].numpy(), size=15)

plt.show()
```

```python
mnist_dataset = torchvision.datasets.MNIST(
    image_path, 'train', download=True
)
```

```python
assert isinstance(mnist_dataset, torch.utils.data.Dataset)
```

```python
example = next(enumerate(mnist_dataset))
print(example)
```

```python
fig = plt.figure(figsize=(15, 10))

for i, (img, label) in islice(enumerate(mnist_dataset), 18):
    ax = fig.add_subplot(3, 6, i+1)
    ax.imshow(img)
    ax.set_title(label, size=15)

plt.show()
```
